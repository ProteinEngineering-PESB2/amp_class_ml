```python
import pandas as pd
import os
```

```python
def collected_models(path_dir):
    list_elements = os.listdir(path_dir)

    list_df = []

    for element in list_elements:
        df = pd.read_csv(f"{path_dir}{element}")
        df["combination"] = element.split(".")[0]
        list_df.append(df)
    
    df_tuned_models = pd.concat(list_df, axis=0)
    return df_tuned_models
```

### Checking antimicrobial

```python
path_amp = "../../optimization_models/antimicrobial/"
tuned_models_amp = collected_models(path_amp)
tuned_models_amp.sort_values(by="test_f1", ascending=False)
```
